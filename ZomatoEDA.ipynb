```python
ratings= final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0:'Rating count'})
```

rating -> 4.5 to 4.9 ---> Excellent

rating -> 4.0 to 4.4 ---> Very Good

rating -> 3.5 to 3.9 ---> Good

rating -> 2.5 to 3.4 ---> Average

rating -> 1.8 to 2.4 ---> Poor

```python
ratings.head()
```

```python
sns.barplot(x="Aggregate rating", y = "Rating count", data=ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating", y = "Rating count", data=ratings)
```

```python
sns.barplot(x="Aggregate rating", hue = 'Rating color', y = "Rating count", data=ratings) #changing into colors
```

```python
sns.barplot(x="Aggregate rating", hue = 'Rating color', y = "Rating count", data=ratings, palette = ['grey', 'red', 'orange', 'yellow', 'green', 'purple']) #changing into right colors using palette
```

Observation

Max rating are between 2.5 to 3.9

```python
sns.countplot(x='Rating color', data = ratings, palette=['grey', 'red', 'orange', 'yellow', 'green', 'purple']) #this gives us the frequency of color present in the data, NOT THE RATING COUNT
```
